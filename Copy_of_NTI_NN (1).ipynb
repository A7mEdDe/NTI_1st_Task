# Binary Cls

## Data

```python
import pandas as pd
from keras import Sequential
from keras.layers import Dense, Input
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("/content/titanic_data.csv")
data.head()
```

### Data Cleaning

```python
data = data.dropna()
data['sex'] = data['sex'].map({'male': 1, 'female': 0})
data['adult_male'] = data['adult_male'].map({True: 1, False: 0})
data['alone'] = data['alone'].map({True: 1, False: 0})
data['class'] = data['class'].map({'First': 1, 'Second': 2, 'Third': 3})
data['who'] = data['who'].map({'man': 1, 'woman': 2, 'child': 3})
data = data.drop(['embarked', 'deck', 'embark_town', 'alive'], axis=1)
data
```

```python
X_train = data.drop("survived", axis=1)
y_train = data['survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
print("the shape of training set : ", X_train.shape)
print("the shape of testing set : ", X_test.shape)
```

## Model

- input layer neuorans ( 10 )
---
- output layer activation function ( sigmoid )
- output layer Loss Function ( adam )
- output layer neourans ( 2 )
---
- hidden layers number ( 3 )
- activation on the hidden layers   ( relu )
- nuorans on hidden layers  ( 10 )

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
```

```python
model = Sequential([
    Input(shape=(10,)),
    Dense(10, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer=Adam(), loss=BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test),epochs=100, batch_size=32)
```

```python
print(model.evaluate(X_test, y_test))
```

# Multi Cls

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Input
import numpy as np
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
data.head()
```

```python
x = data.drop("target", axis=1)
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = Sequential(
    [
        Input(shape=(4,)),
        Dense(10, activation='relu'),
        Dense(5, activation='relu'),
        Dense(3, activation='softmax')
    ]
)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test),epochs=50, batch_size=32)
```

```python
accuracy = model.evaluate(X_test, y_test)
print(f"Accuracy : {accuracy[1] * 100:.2f} %")
```
